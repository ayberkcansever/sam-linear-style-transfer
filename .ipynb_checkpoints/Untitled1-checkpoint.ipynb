```python
import pandas as pd
import psycopg2
import numpy as np
import statsmodels.api as sm
import math

conn = psycopg2.connect(database="crypto",
                        host="localhost",
                        user="postgres",
                        password="postgres",
                        port="6432")
cursor = conn.cursor()

query = ("""SELECT moment, high, open, low, close, volume FROM kline_15 where crypto_name ='ETHTRY' 
and to_timestamp(moment / 1000) > '2023-01-01'
and to_timestamp(moment / 1000) < '2024-01-01'
ORDER BY moment""")

sql_query = pd.read_sql(query, conn)
df = pd.DataFrame(sql_query)

df['moment'] = pd.to_datetime(df['moment'], unit='ms')

# Calculate Rolling Averages
df.set_index('moment', inplace=True)

df['1h_rolling_avg'] = df['close'].rolling(window=4, min_periods=1).mean()
df['24h_rolling_avg'] = df['close'].rolling(window=96, min_periods=1).mean()
df['1h_pct_change'] = (df['1h_rolling_avg'] - df['1h_rolling_avg'].shift(4)) / df['1h_rolling_avg'].shift(4) * 100
df['1d_pct_change'] = (df['24h_rolling_avg'] - df['24h_rolling_avg'].shift(4*24*1)) / df['24h_rolling_avg'].shift(4*24*1) * 100
df['7d_pct_change'] = (df['24h_rolling_avg'] - df['24h_rolling_avg'].shift(4*24*7)) / df['24h_rolling_avg'].shift(4*24*7) * 100
df['15d_pct_change'] = (df['24h_rolling_avg'] - df['24h_rolling_avg'].shift(4*24*15)) / df['24h_rolling_avg'].shift(4*24*15) * 100
df['30d_pct_change'] = (df['24h_rolling_avg'] - df['24h_rolling_avg'].shift(4*24*30)) / df['24h_rolling_avg'].shift(4*24*30) * 100
df['30d_pct_change_diff'] = df['30d_pct_change'] - df['30d_pct_change'].shift(4*24)

df.reset_index(inplace=True)

# Calculate Bollinger Bands
window = 96
df['SMA'] = df['close'].rolling(window=window).mean()
df['SD'] = df['close'].rolling(window=window).std()
df['Upper_BB'] = df['SMA'] + (2 * df['SD'])
df['Lower_BB'] = df['SMA'] - (2 * df['SD'])
df['Width_BB'] = df['Upper_BB'] - df['Lower_BB']
df['Lower_diff'] = df['close'] - df['Lower_BB']
df['Upper_diff'] = df['close'] - df['Upper_BB']
df['Lower_diff_pct'] = (df['close'] - df['Lower_BB']) / (df['Upper_BB'] - df['Lower_BB'])
df['Upper_diff_pct'] = (df['close'] - df['Upper_BB']) / (df['Upper_BB'] - df['Lower_BB'])

# Calculate RSI
rsi_window = 24
df['delta'] = df['close'].diff()
df['gain'] = df['delta'].clip(lower=0)
df['loss'] = -df['delta'].clip(upper=0)
df['avg_gain'] = df['gain'].rolling(window=rsi_window, min_periods=1).mean()
df['avg_loss'] = df['loss'].rolling(window=rsi_window, min_periods=1).mean()
df['rs'] = df['avg_gain'] / df['avg_loss']
df['rsi'] = 100 - (100 / (1 + df['rs']))

# Calculate ATR
def calculate_atr(dataframe, period=96):
    dataframe['high_low'] = dataframe['high'] - dataframe['low']
    dataframe['high_close'] = (dataframe['high'] - dataframe['close'].shift()).abs()
    dataframe['low_close'] = (dataframe['low'] - dataframe['close'].shift()).abs()
    dataframe['tr'] = dataframe[['high_low', 'high_close', 'low_close']].max(axis=1)
    atr = dataframe['tr'].rolling(window=period).mean()
    return atr

atr_values = calculate_atr(df, period=24)
df['ATR'] = atr_values

df.to_csv('ETH_2022_2023_2024.csv', index=False)
```

```python
import gym
from gym import spaces
import numpy as np
from gym.utils import seeding
from sklearn.preprocessing import MinMaxScaler
import warnings
from prettytable import PrettyTable
import tensorflow as tf
import joblib
from datetime import datetime

class TradingEnv(gym.Env):
    """A trading environment for reinforcement learning."""
    metadata = {'render.modes': ['human']}


    def __init__(self, df, scaler=None, enable_log=False):
        super(TradingEnv, self).__init__()
        self.df = df.reset_index()
        self.enable_log = enable_log
        self.episode_number = 1
        
        self.table = PrettyTable()
        self.table.field_names = ["Act.", "Moment", "Pr", "B.Pr", "Rew", "Prf", "P%", "LB", "UB", "RSI"]
        self.table.align = "r"
        
        train_log_dir = '/Users/ayberk.cansever/Documents/ECU/Thesis/SAM/ppo_trading_logs/'
        self.train_summary_writer = tf.summary.create_file_writer(train_log_dir)
        
        if scaler is None:
            self.scaler = MinMaxScaler()
            self.scaler.fit(self.df[['high', 'open', 'low', 'close', 'Lower_diff', 'Upper_diff', 'Width_BB', 'rsi']])
            joblib.dump(self.scaler, 'scaler.pkl')
            print('scaler is created.')
        else:
            self.scaler = scaler
            print('scaler is being used.')
        
        self.observation_space = spaces.Box(low=0, high=1, shape=(len(df.columns) - 1,), dtype=np.float32)
        self.reward_range = (-1, 1)
        self.action_space = spaces.Discrete(3)  # 0: hold, 1: buy, 2: sell
        
        self.current_step = 0
        self.total_profit = 0
        self.done = False
        self.current_position = None
        self.last_buy_price = None
        self.commission_rate = 0.003

    def _next_observation(self):
        obs = self.df.loc[self.current_step, ['high', 'open', 'low', 'close', 'Lower_diff', 'Upper_diff', 'Width_BB', 'rsi']].to_numpy()
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            scaled_obs = self.scaler.transform([obs])
        return scaled_obs

    def step(self, action):
        self.current_step += 1

        if self.current_step >= len(self.df) - 1:
            self.done = True
            with self.train_summary_writer.as_default():
                tf.summary.scalar('total_profit', self.total_profit, step=self.episode_number)
                self.train_summary_writer.flush()
            
        moment = self.df.loc[self.current_step, 'moment']
        current_price = self.df.loc[self.current_step, 'close']
        lower_diff = self.df.loc[self.current_step, 'Lower_diff']
        upper_diff = self.df.loc[self.current_step, 'Upper_diff']
        rsi = self.df.loc[self.current_step, 'rsi']
        
        reward = 0
        
        
        if action == 0:  # Hold
            reward -= 0.5
            
            # Bollinger Bands - Buy
            if lower_diff < 0:
                reward -= 0.5
            
            # Bollinger Bands - Sell
            if upper_diff > 0:
                reward -= 0.5
            
            if self.enable_log:
                self.table.add_row(["Hold", moment, f"{current_price:.2f}", f"{current_price:.2f}", f"{reward:.2f}", 0, 0, f"{lower_diff:.1f}", f"{upper_diff:.1f}", f"{rsi:.0f}"])
           
        
        elif action == 1:  # Buy
            # Bollinger Bands - Buy
            if lower_diff <= 0:
                reward += 1.0
            # Bollinger Bands - Sell
            elif upper_diff >= 0:
                reward -= 1.0
            else:
                reward -= 1.0
            
            if self.enable_log:
                self.table.add_row(["Buy", moment, f"{current_price:.2f}", f"{current_price:.2f}", f"{reward:.2f}", 0, 0, f"{lower_diff:.1f}", f"{upper_diff:.1f}", f"{rsi:.0f}"])
            
            
        elif action == 2:  # Sell
            
            # Bollinger Bands - Sell
            if upper_diff >= 0:
                reward += 1.0
                
            # Bollinger Bands - Buy
            elif lower_diff <= 0:
                reward -= 1.0
            else:
                reward -= 1.0

            #commission = (current_price * self.commission_rate)
            #profit = current_price - commission - self.last_buy_price
            #self.total_profit += profit
            #profit_percentage = profit / self.last_buy_price

            if self.enable_log:
                self.table.add_row(["Sell", moment, f"{current_price:.2f}", 0, f"{reward:.2f}", 0, 0, f"{lower_diff:.1f}", f"{upper_diff:.1f}", f"{rsi:.0f}"])

        return self._next_observation(), reward, self.done, {}
    
    def seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]

    def reset(self):
        self.current_step = 0
        self.total_profit = 0
        self.done = False
        self.current_position = None
        self.last_buy_price = None
        self.episode_number = 1
        return self._next_observation()

    def render(self, mode='human'):
        if self.enable_log:
            print(self.table)
            print(f'Profit: {self.total_profit}')
        
```

```python
from stable_baselines3 import DQN, PPO, A2C, SAC, TD3
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
import pandas as pd

# Initialize the environment
env = TradingEnv(df)

# Initialize the agent
#model = DQN("MlpPolicy", vec_env, verbose=0, tensorboard_log="./ppo_trading_logs/")
model = PPO("MlpPolicy", env, verbose=0, tensorboard_log="./ppo_trading_logs/")
#model = A2C("MlpPolicy", vec_env, verbose=0, tensorboard_log="./ppo_trading_logs/")
#model = SAC("MlpPolicy", vec_env, verbose=0)
#model = TD3("MlpPolicy", vec_env, verbose=0)

# Train the agent
model.learn(total_timesteps=1000000)

# Save the model
model.save("ac_crypto_trading_model")

env.render()
```

```python
import datetime
import pandas as pd
import psycopg2
import numpy as np
import statsmodels.api as sm
import math

def evaluate_model(model, validation_env, num_episodes=1):
    episode_rewards = []
    episode_lengths = []

    for i in range(num_episodes):
        obs = validation_env.reset()
        done = False
        total_rewards = 0.0
        steps = 0

        while not done:
            action, _states = model.predict(obs, deterministic=False)
            obs, reward, done, info = validation_env.step(action)
            total_rewards += reward
            steps += 1
        
        episode_rewards.append(total_rewards)
        episode_lengths.append(steps)

    mean_reward = np.mean(episode_rewards)
    std_reward = np.std(episode_rewards)
    mean_length = np.mean(episode_lengths)

    print(f"Mean reward: {mean_reward:.2f} +/- {std_reward:.2f}")
    print(f"Mean episode length: {mean_length:.2f}")

    return episode_rewards

conn = psycopg2.connect(database="crypto",
                        host="localhost",
                        user="postgres",
                        password="postgres",
                        port="6432")
cursor = conn.cursor()

query = ("""SELECT moment, high, open, low, close, volume FROM kline_15 where crypto_name ='ETHTRY' 
and to_timestamp(moment / 1000) >= '2022-01-01'
ORDER BY moment""")

#query = ("""SELECT moment, close FROM kline where crypto_name ='ADATRY' ORDER BY moment""")

sql_query = pd.read_sql(query, conn)
df_val = pd.DataFrame(sql_query)

# Convert 'moment' from UTC milliseconds to datetime
df_val['moment'] = pd.to_datetime(df_val['moment'], unit='ms')
df_val.set_index('moment', inplace=True)
df_val.reset_index(inplace=True)

# Calculate Bollinger Bands
window = 96
df_val['SMA'] = df_val['close'].rolling(window=window).mean()
df_val['SD'] = df_val['close'].rolling(window=window).std()
df_val['Upper_BB'] = df_val['SMA'] + (2 * df_val['SD'])
df_val['Lower_BB'] = df_val['SMA'] - (2 * df_val['SD'])
df_val['Width_BB'] = df_val['Upper_BB'] - df_val['Lower_BB']
df_val['Lower_diff'] = df_val['close'] - df_val['Lower_BB']
df_val['Upper_diff'] = df_val['close'] - df_val['Upper_BB']
df_val['Lower_diff_pct'] = (df_val['close'] - df_val['Lower_BB']) / (df_val['Upper_BB'] - df_val['Lower_BB'])
df_val['Upper_diff_pct'] = (df_val['close'] - df_val['Upper_BB']) / (df_val['Upper_BB'] - df_val['Lower_BB'])

# Calculate RSI
rsi_window = 24
df_val['delta'] = df_val['close'].diff()
df_val['gain'] = df_val['delta'].clip(lower=0)
df_val['loss'] = -df_val['delta'].clip(upper=0)
df_val['avg_gain'] = df_val['gain'].rolling(window=rsi_window, min_periods=1).mean()
df_val['avg_loss'] = df_val['loss'].rolling(window=rsi_window, min_periods=1).mean()
df_val['rs'] = df_val['avg_gain'] / df_val['avg_loss']
df_val['rsi'] = 100 - (100 / (1 + df_val['rs']))


df_val = df_val[['moment', 'high', 'open', 'low', 'close', 'Lower_diff', 'Upper_diff', 'Width_BB', 'rsi']]
df_val = df_val.dropna()

df_val = df_val[(df_val['moment'] > '2022-03-01 00:00:00') & (df_val['moment'] < '2022-03-10 00:00:00')]

scaler = joblib.load('scaler.pkl')
validation_env = TradingEnv(df_val, scaler, True)

# Evaluate the model
trained_model = model.load("ac_crypto_trading_model")
evaluate_model(trained_model, validation_env, num_episodes=1)
validation_env.render()
```
