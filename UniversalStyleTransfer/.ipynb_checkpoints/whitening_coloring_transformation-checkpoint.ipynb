```python
import numpy as np
import cv2
import torch
import torch.linalg as LA
import matplotlib.pyplot as plt

def image_to_tensor(image):
    """Convert a BGR image to a PyTorch tensor and normalize it to range [0, 1]."""
    image = image.astype(np.float32) / 255.0  # Scale to [0, 1]
    image = np.transpose(image, (2, 0, 1))  # Convert HWC to CHW
    return torch.from_numpy(image).unsqueeze(0)  # Add batch dimension

def tensor_to_image(tensor):
    """Convert a PyTorch tensor back to a BGR image and clamp pixel values."""
    tensor = tensor.squeeze(0).detach().cpu().numpy()  # Remove batch dimension
    image = np.transpose(tensor, (1, 2, 0))  # Convert CHW to HWC
    image = np.clip(image * 255, 0, 255).astype(np.uint8)  # Ensure valid pixel range [0, 255]
    return image

def compute_covariance(matrix):
    """Compute the covariance matrix of the input matrix."""
    return matrix @ matrix.T / matrix.size(1)

def whitening(content_features, epsilon=1e-5):
    """Apply whitening to the content features using SVD for stability."""
    # Compute mean and subtract it
    c_mean = torch.mean(content_features, dim=1, keepdim=True)
    c_centered = content_features - c_mean

    # SVD (Singular Value Decomposition) instead of eigenvalue decomposition
    U, S, V = torch.svd(compute_covariance(c_centered))

    # Add epsilon to singular values to prevent division by zero
    S = torch.clamp(S, min=epsilon)

    # Whitening transformation
    diag_inv_sqrt = torch.diag(torch.rsqrt(S))
    whitened_features = U @ diag_inv_sqrt @ U.T @ c_centered

    return whitened_features, c_mean

def coloring(whitened_features, style_mean, style_features, epsilon=1e-5):
    """Apply coloring to the whitened features using SVD for stability."""
    # Compute covariance matrix of style features
    cov_s = compute_covariance(style_features)

    # SVD for coloring
    U_s, S_s, V_s = torch.svd(cov_s)

    # Add epsilon to singular values to prevent instability
    S_s = torch.clamp(S_s, min=epsilon)

    # Coloring transformation
    diag_s_sqrt = torch.diag(torch.sqrt(S_s))
    colored_features = U_s @ diag_s_sqrt @ U_s.T @ whitened_features

    # Add back the style mean (which is same as the content mean here)
    colored_features += style_mean

    return colored_features

def whitening_and_coloring(image):
    """Apply whitening and coloring transform to the same image."""
    # Convert image to tensor
    content_tensor = image_to_tensor(image)
    style_tensor = content_tensor.clone()  # Style is the same as content

    # Reshape to (channels, spatial_size)
    content_features = content_tensor.view(content_tensor.size(1), -1)
    style_features = style_tensor.view(style_tensor.size(1), -1)

    # Step 1: Whitening
    whitened_features, content_mean = whitening(content_features)

    # Step 2: Coloring (same image as style)
    colored_features = coloring(whitened_features, content_mean, style_features)

    # Reshape back to original dimensions
    colored_tensor = colored_features.view_as(content_tensor)

    # Convert tensor back to image
    reconstructed_image = tensor_to_image(colored_tensor)

    return reconstructed_image

# Load your image (make sure it's in BGR format as OpenCV uses BGR)
image_path = '/Users/ayberk.cansever/Documents/ECU/Thesis/SAM/sources/image-4.png'
original_image = cv2.imread(image_path)

# Apply whitening and coloring directly
reconstructed_image = whitening_and_coloring(original_image)

# Plot the original and reconstructed image
plt.figure(figsize=(10, 5))

# Plot original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
plt.title('Original Image')
plt.axis('off')

# Plot reconstructed image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(reconstructed_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
plt.title('Reconstructed Image')
plt.axis('off')

plt.show()
```
